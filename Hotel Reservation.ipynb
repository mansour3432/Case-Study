# Hotel Reservation Dataset 

>**The Data from [Kaggle](https://www.kaggle.com/datasets/ahsan81/hotel-reservations-classification-dataset)**

**description of each column:**

- Booking_ID (col_character): Unique identifier for each booking.
- no_of_adults (col_double): Number of adults included in the booking.
- no_of_children (col_double): Number of children included in the booking.
- no_of_weekend_nights (col_double): Number of nights reserved for the weekend.
- no_of_week_nights (col_double): Number of nights reserved for weekdays.
- type_of_meal_plan (col_character): Type of meal plan selected for the booking.
- required_car_parking_space (col_double): Indicates whether a car parking space was requested.
- room_type_reserved (col_character): Type of room reserved.
- lead_time (col_double): Number of days between booking and arrival.
- arrival_year (col_double): Year of arrival.
- arrival_month (col_double): Month of arrival.
- arrival_date (col_double): Date of arrival.
- market_segment_type (col_character): The market segment associated with the booking.
- repeated_guest (col_double): Indicates whether the guest is a repeated visitor (1 for repeated, 0 for first-time).
- no_of_previous_cancellations (col_double): Number of previous cancellations by the guest.
- no_of_previous_bookings_not_canceled (col_double): Number of previous bookings that were not canceled by the guest.
- avg_price_per_room (col_double): Average price per room for the booking.
- no_of_special_requests (col_double): Number of special requests made by the guest.
- booking_status (col_character): The status of the booking (e.g., confirmed, canceled).

.

>## **Data Cleaning & Manipulation**

.

```r
# rename the arrival_ date column to arrival_day_of_month.
data <- data %>% 
    rename(arrival_day_of_month= arrival_date )
```

```r
# add arrival_date to the dataset by combining the three column in one.
data <- data%>% 
    mutate(arrival_date = as.Date(paste(arrival_year, arrival_month, arrival_day_of_month, sep ="-")))
```

```r
# Extract day labels (e.g., "Mon", "Tue", ...) and save it in new column.
data <- data %>%
  mutate(day_of_week = format(arrival_date, "%a"))
```

```r
# count the NULL values.
colSums(is.na(data))
```

```r
# show the NULL values.
null_values <- data %>% 
    select(arrival_date, day_of_week, arrival_year, arrival_day_of_month, arrival_month) %>% 
    filter(is.na(arrival_date))

null_values
```

*In 2018 the february Month was only 28 days*

*Drop the null rows*

```r
data <- data %>%  
    filter(!is.na(arrival_date))
```

.

```r
# add arrival_month_name column by extract the month name from the arrival month column
data<- data %>% 
    mutate(arrival_month_name = month.name[arrival_month])
```
